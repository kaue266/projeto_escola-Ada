```python
import json
import csv

import os # Usado para verificar a existencia da base de alunos
from copy import deepcopy # usado para poder criar copia do dict sem erros

from rich.console import Console# painel bonitinho
from rich.panel import Panel
from rich.table import Table
from rich.text import Text

import pandas as pd # para baixar boletim

from datetime import datetime
from datetime import date # usado para calcular a idade

console = Console()

#template dos dados
alunos = {
    "0" :{"nome": "",
    "data_nascimento": "",
    "contato": {
        "telefone": "",
        "email": ""
    },
    "turma": {
        "serie": "",
        "turno": ""
    },
    "trimestre_1": [
        {
            "disciplina": "portugues",
            "provas": []

        },
        {
            "disciplina": "matematica",
            "provas": []

        },
        {
            "disciplina": "ciencias",
            "provas": []

        },
        {
            "disciplina": "historia",
            "provas": []

        },
        {
            "disciplina": "geografia",
            "provas": []

        },
        {
            "disciplina": "educacao_fisica",
            "provas": []

        },
        {
            "disciplina": "ingles",
            "provas": []

        },
        {
            "disciplina": "artes",
            "provas": []
        }
    ],
    "trimestre_2": [
        {
            "disciplina": "portugues",
            "provas": []

        },
        {
            "disciplina": "matematica",
            "provas": []

        },
        {
            "disciplina": "ciencias",
            "provas": []

        },
        {
            "disciplina": "historia",
            "provas": []

        },
        {
            "disciplina": "geografia",
            "provas": []

        },
        {
            "disciplina": "educacao_fisica",
            "provas": []

        },
        {
            "disciplina": "ingles",
            "provas": []
        },
        {
            "disciplina": "artes",
            "provas": []

        }],
    "trimestre_3": [
        {
            "disciplina": "portugues",
            "provas": []

        },
        {
            "disciplina": "matematica",
            "provas": []

        },
        {
            "disciplina": "ciencias",
            "provas": []

        },
        {
            "disciplina": "historia",
            "provas": []
        },
        {
            "disciplina": "geografia",
            "provas": []

        },
        {
            "disciplina": "educacao_fisica",
            "provas": []

        },
        {
            "disciplina": "ingles",
            "provas": []

        },
        {
            "disciplina": "artes",
            "provas": []

        }]
}}

def abre_dados( tipo:str ) -> dict :
    '''Abre o arquivo com o tipo de operação desejada ('r' para leitura)'''
    with open('alunos.json', tipo) as dados:
        alunos = json.load(dados)
    return alunos

def verifica_matricula(matricula):
      alunos = abre_dados("r")
      if matricula not in alunos:
          console.print(f"Matrícula {matricula} não encontrada.")
          return False
      else:
       return True

def matricula_funcao(tipo:str) -> int:
    ''' Funcão que verifica ou gera matricula.
    comandos: "V" para verificar alunos e 'G' para gerar novo aluno '''

    alunos = abre_dados('r')
    matriculas = list(alunos.keys()) # Lista de chaves, que são as matrículas

    # verifica qual vai ser a função.
    if tipo.lower() == "v":
        matriculas.remove("0")
        nova_matricula = len(matriculas)
        return nova_matricula

    elif tipo.lower() == "g":
        nova_matricula = int(max(matriculas)) + 1
        return nova_matricula

    else:
        console.print("Comando entregado para a função não existe")

def inicializa_verificacao():
  ''' Vefifica a base de alunos em arquivo '''
  if os.path.isfile('alunos.json'):
      console.print("Base de dados existe")
  else:
      # cria e abre o aquivo com os dados do aluno limpo.
      with open('alunos.json', 'w') as arquivo:
        json.dump(alunos, arquivo, indent=4)
        console.print(" Base de dados criada com sucesso!")

def adiciona_dados(alunos):
    matricula = matricula_funcao("G")
    console.print(f"Nova matrícula: {matricula}")

    try:
        nome = console.input("Digite o nome: ")
        serie = console.input("Digite a série: ")
        turno = console.input("Digite o turno (Matutino/Vespertino/Noturno): ")
        if turno.lower() not in ["matutino", "vespertino", "noturno"]:
            raise ValueError("Turno inválido. Escolha entre Matutino, Vespertino ou Noturno.")

        data_nascimento = console.input("Digite a data de nascimento (DD/MM/AAAA): ")

        try:
         data_nascimento = datetime.strptime(data_nascimento, "%d/%m/%Y")
         data_nascimento_mod = data_nascimento.strftime("%Y-%m-%d")

        except ValueError:
            raise ValueError("Data de nascimento inválida. Deve estar no formato DD/MM/AAAA.")

        telefone = console.input("Digite o número de telefone (somente números): ")
        if not telefone.isdigit():
            raise ValueError("Telefone deve conter apenas números.")

        email = console.input("Digite o email: ")
        if "@" not in email or "." not in email:
            raise ValueError("Email inválido. Deve conter '@' e um domínio.")

    except ValueError as e:
        console.print(f"[red]Erro de entrada: {e}")
        return
    # concatena serie
    serie = serie + "° ano"

    novo_aluno = deepcopy(alunos["0"])
    novo_aluno["nome"] = nome
    novo_aluno["data_nascimento"] = data_nascimento_mod
    novo_aluno["contato"]["telefone"] = telefone
    novo_aluno["contato"]["email"] = email
    novo_aluno["turma"]["serie"] = serie
    novo_aluno["turma"]["turno"] = turno.lower()

    alunos = abre_dados('r')
    # junta tudo com os dados e a matrícula
    alunos[matricula] = novo_aluno

    # salva tudo
    with open('alunos.json', 'w') as arquivo:
        json.dump(alunos, arquivo, indent=4)
    console.print(f"Aluno {nome} matriculado com sucesso!")

def adiciona_atualiza_nota():
    try:
        numero_matricula = console.input("Digite o número de matrícula: ")
        if not numero_matricula.isdigit():
            raise ValueError("Número de matrícula deve conter apenas números.")

        trimestre = int(console.input("Qual é o trimestre aplicação da prova? (digite o número): "))
        if trimestre not in [1, 2, 3]:
            raise ValueError("Trimestre inválido. Escolha 1, 2 ou 3.")

        materia = console.input("Digite a disciplina: ").lower()
        tipo = console.input("Qual é o tipo da prova (ex: av1, av2, rec): ")
        if tipo not in ["av1", "av2", "rec"]:
            raise ValueError("O tipo de prova é inválido. Escolha av1, av2, rec.")

        nota = float(console.input("Qual foi a nota dada ao aluno?: "))
        if nota < 0 or nota > 10:
            raise ValueError("Nota inválida. Deve estar entre 0 e 10.")

    except ValueError as e:
        console.print(f"[red]Erro de entrada: {e}")
        return

    # Abre os dados
    alunos = abre_dados('r')
    if not verifica_matricula(numero_matricula):
        console.print(f"[red]Matrícula {numero_matricula} não encontrada.[/red]")
        return

    aluno = alunos[numero_matricula]  # Obtém os dados do aluno

    # Analisa se o trimestre já existe ou cria um novo
    if f"trimestre_{trimestre}" not in aluno:
        aluno[f"trimestre_{trimestre}"] = []

    encontrou_materia = False
    encontrou_prova = False

    # Procura a disciplina correta nas avaliações do aluno no trimestre especificado
    for avaliacao in aluno[f"trimestre_{trimestre}"]:
        if avaliacao["disciplina"] == materia:
            # A disciplina foi encontrada, depois procurar o tipo de prova
            for prova in avaliacao["provas"]:
                if prova["tipo"] == tipo:
                    encontrou_prova = True
                    console.print(f"Nota da {tipo} já existe, vai atualizar? ")
                    nova_nota = console.input("Quer atualizar a nota? Digite a nova nota ou 'sair': ")

                    if nova_nota.lower() == "sair":
                        break
                    try:
                        nova_nota = float(nova_nota)
                        if nova_nota < 0 or nova_nota > 10:
                            raise ValueError("Nota inválida. Deve estar entre 0 e 10.")
                        prova["nota"] = nova_nota
                        console.print(f"[green]Nota atualizada para {nova_nota}.[/green]")
                    except ValueError:
                        console.print("[red]Nota inválida. Deve ser um número entre 0 e 10.[/red]")
                        return

            if not encontrou_prova:
                avaliacao["provas"].append({
                    "tipo": tipo,
                    "nota": nota
                })
                console.print(f"[green]Nota da prova {tipo} adicionada com sucesso para a disciplina {materia}.[/green]")
            encontrou_materia = True
            break

    # Se a disciplina não foi encontrada, cria uma nova entrada
    if not encontrou_materia:
        nova_avaliacao = {
            "disciplina": materia,
            "provas": [
                {
                    "tipo": tipo,
                    "nota": nota
                }
            ]
        }
        aluno[f"trimestre_{trimestre}"].append(nova_avaliacao)
        console.print(f"[green]Nova disciplina {materia} adicionada com nota {nota}.[/green]")

    # Atualiza o arquivo JSON com todas as alterações
    with open('alunos.json', 'w') as arquivo:
        json.dump(alunos, arquivo, indent=4)

    console.print("[green]Dados salvos com sucesso![/green]")
    input("Aperte qualquer tecla para continuar **PARE DE PROCURAR A QUALQUER TECLA \n")

def altera_dados():
    try:
        matricula = input("Digite o número de matrícula: ")
        print("Escolha o que deseja alterar: 1. Nome 2. Data de nascimento 3. Turma 4. Contato")
        dados_alterar = int(input("Qual dado deseja alterar? "))

        if dados_alterar == 1:
            dados_alterar = "nome"
        elif dados_alterar == 2:
            dados_alterar = "data_nascimento"
        elif dados_alterar == 3:
            dados_alterar = "turma"
        elif dados_alterar == 4:
            dados_alterar = "contato"
        else:
            print("Opção incorreta!!")
            return

        dados = abre_dados('r')

        # Verifica qual o tipo de dados vai ser alterado
        if dados_alterar == "contato":
            contato = input("Deseja alterar 'telefone' ou 'email'? ").lower()
            if contato == "telefone":
                telefone_atual = dados[matricula]["contato"]["telefone"]
                print(f"Telefone cadastrado: {telefone_atual}")
                novo_telefone = input("Digite o novo telefone: ")
                dados[matricula]["contato"]["telefone"] = novo_telefone
                print(f"Telefone atualizado com sucesso: {novo_telefone}")
            elif contato == "email":
                email_atual = dados[matricula]["contato"]["email"]
                print(f"E-mail cadastrado: {email_atual}")
                novo_email = input("Digite o novo e-mail: ")
                dados[matricula]["contato"]["email"] = novo_email
                print(f"E-mail atualizado com sucesso: {novo_email}")
            else:
                print("Opção de contato inválida.")
        elif dados_alterar == "turma":
            serie_atual = dados[matricula]["turma"]["serie"]
            turno_atual = dados[matricula]["turma"]["turno"]
            print(f"Série atual: {serie_atual}, Turno atual: {turno_atual}")
            nova_serie = input("Digite a nova série: ")
            nova_serie = nova_serie +"° ano"
            novo_turno = input("Digite o novo turno: ")
            dados[matricula]["turma"].update({"serie": nova_serie, "turno": novo_turno})
            print(f"Dados da turma atualizados para série: {nova_serie}, turno: {novo_turno}")

        elif dados_alterar == "nome":
            nome_atual = dados[matricula]["nome"]
            print(f"Nome atual: {nome_atual}")
            novo_nome = input("Digite o novo nome: ")
            dados[matricula]["nome"] = novo_nome
            print(f"Nome atualizado com sucesso para {novo_nome}.")

        elif dados_alterar == "data_nascimento":
            data_atual = dados[matricula]["data_nascimento"]
            print(f"Data de nascimento atual: {data_atual}")
            nova_data = input("Digite a nova data de nascimento (DD/MM/AAAA): ")

            # Tenta converter a nova data para garantir que está no formato correto
            try:
                nova_data_obj = datetime.strptime(nova_data, "%d/%m/%Y").strftime("%Y-%m-%d")
                dados[matricula]["data_nascimento"] = nova_data_obj
                print(f"Data de nascimento atualizada para {nova_data_obj}.")
            except ValueError:
                print("Formato de data inválido! Por favor, use o formato DD/MM/AAAA.")
                return  # Sai da função se o formato for inválido

        # Salva as alterações no arquivo JSON
        try:
            with open('alunos.json', 'w') as arquivo:
                json.dump(dados, arquivo, indent=4)
            print("Alterações salvas com sucesso.")
        except Exception as e:
            print(f"Erro ao salvar as alterações: {e}")

    except ValueError as e:
        print(f"Erro na entrada: {e}")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

    input("Aperte qualquer tecla para continuar **PARE DE PROCURAR A QUALQUER TECLA \n")


def remove_aluno():
    try:
        matricula = console.input("Digite o número de matrícula: ")
        dados = abre_dados("r")

        if matricula == "0":
            console.print("[red]Você não pode apagar o aluno que não existe.")
            return

        if verifica_matricula(matricula):
            nome = dados[matricula]["nome"]
            confirmacao = console.input(f"Tem certeza que deseja remover o aluno {nome}? (S/N): ").lower()

            if confirmacao == 's':
                del dados[matricula]

                with open('alunos.json', 'w') as arquivo:
                    json.dump(dados, arquivo, indent=4)

                console.print(f"[green]Aluno {nome} removido com sucesso.")
            else:
                console.print("[yellow]Operação cancelada.")
        else:
            console.print("[red]Matrícula não encontrada.")

    except FileNotFoundError:
        console.print("[red]Arquivo 'alunos.json' não encontrado.")

    except KeyError:
        console.print("[red]Erro ao acessar dados do aluno.")

    except Exception as e:
        console.print(f"[red]Ocorreu um erro inesperado: {e}")

    input("Aperte qualquer tecla para continuar **PARE DE PROCURAR A QUALQUER TECLA \n")


def dados_alunos():
    contagem_series = {}
    contagem_turmas = {}
    contagem_turnos = {"matutino": 0, "noturno": 0}

    console.print(f"temos o total de {matricula_funcao('V')}")

    # Verifica se os alunos por turma
    for matricula, dados_aluno in alunos.items():
        serie = dados_aluno["turma"]["serie"]
        turno = dados_aluno["turma"]["turno"]

        # Contagem por série
        if serie not in contagem_series:
            contagem_series[serie] = 1
        else:
            contagem_series[serie] += 1

        # Contagem por turma (considerando a série como referência de turma)
        if serie not in contagem_turmas:
            contagem_turmas[serie] = 1
        else:
            contagem_turmas[serie] += 1


    # Exibe os resultados
    console.print("\nContagem por Série:")
    for serie, count in contagem_series.items():
        console.print(f"Série {serie.capitalize()}: {count} alunos")

    console.print("\nContagem por Turma:")
    for turma, count in contagem_turmas.items():
        console.print(f"Turma {turma.capitalize()}: {count} alunos")

    console.print("\nContagem por Turno:")
    for turno, count in contagem_turnos.items():
        console.print(f"Turno {turno.capitalize()}: {count} alunos")

    input("Aperte qualquer tecla para continuar **PARE DE PROCURAR A QUALQUER TECLA \n")

def exibir_boletim():
    """Calcula a média e a situação de um aluno para uma disciplina específica."""

    matricula = console.input("Digite o numero da matricula: ")
    disciplina = console.input("Digite a materia: ")
    nota_min_aprov = int(console.input("Digite a nota minima de aprovação: "))

    dados_alunos = abre_dados('r')

    if matricula not in dados_alunos:
        return "Matrícula não encontrada."
    aluno = dados_alunos[matricula]
    notas_totais = []

    for trimestre, disciplinas in aluno['trimestres'].items():
        for materia in disciplinas:
            if materia['disciplina'] == disciplina:
                notas = [prova['nota'] for prova in materia['provas']]
                notas_totais.extend(notas)

    if not notas_totais:
        return "Nenhuma prova registrada para esta disciplina."

    media = sum(notas_totais) / len(notas_totais)
    situacao = "APROVADO" if media >= nota_min_aprov else "REPROVADO"

    return media, situacao

def exibir_boletim_completo():
    """Mostra todas as notas do aluno em uma tabela e permite baixar como CSV."""

    matricula = console.input('Digite o numero de matricula')

    dados_alunos = abre_dados('r')
    if matricula not in dados_alunos:
        console.print("Matrícula não encontrada.", style="bold red")
        return

    aluno = dados_alunos[matricula]

    notas_list = []

    for trimestre in ['trimestre_1', 'trimestre_2', 'trimestre_3']:
        disciplinas = aluno[trimestre]
        for materia in disciplinas:
            disciplina = materia['disciplina']
            notas = [prova['nota'] for prova in materia['provas']]
            if notas:
                for nota in notas:
                    notas_list.append({
                        "Disciplina": disciplina,
                        "Nota": nota,
                        "Trimestre": trimestre
                    })

    if not notas_list:
        console.print("Nenhuma nota registrada para este aluno.", style="bold yellow")
        return

    tabela = Table(title=f"Notas de {aluno['nome']}", show_header=True, header_style="bold magenta")
    tabela.add_column("Disciplina", style="dim")
    tabela.add_column("Nota", justify="right")
    tabela.add_column("Trimestre", justify="center")

    for nota in notas_list:
        tabela.add_row(nota["Disciplina"], f"{nota['Nota']:.2f}", nota["Trimestre"])

    console.print(tabela)

    if input("Deseja baixar as notas como CSV? Digite sim:").lower() == "sim":
        nome_arquivo = f"notas_{matricula}.csv"
        with open(nome_arquivo, mode='w', newline='') as arquivo_csv:
            writer = csv.DictWriter(arquivo_csv, fieldnames=["Disciplina", "Nota", "Trimestre"])
            writer.writeheader()
            writer.writerows(notas_list)
        console.print(f"Tabela salva como {nome_arquivo}.", style="bold green")

def visualiza_dados_aluno():
    '''Exibe todos os dados dos alunos cadastrados'''
    try:
      alunos = abre_dados('r')
      matricula = console.input("Digite o numero da matricula: ")
      if verifica_matricula(matricula) == True:
            info = alunos[matricula]
            console.print(f"\n[bold]Matrícula:[/] {matricula}")
            console.print(f"[bold]Nome:[/] {info.get('nome', 'Não informado')}")
            console.print(f"[bold]Data de Nascimento:[/] {info.get('data_nascimento', 'Não informado')}")
            calculo_idade = info.get('data_nascimento', '0')
            data_atual = datetime.now().date()
            data_nascimento = datetime.strptime(calculo_idade, "%Y-%m-%d").date()
            idade = data_atual.year - data_nascimento.year - ((data_atual.month, data_atual.day) < (data_nascimento.month, data_nascimento.day))
            console.print(f"[bold]Idade:[/] {idade}")
            console.print(f"[bold]Telefone:[/] {info['contato'].get('telefone', 'Não informado')}")
            console.print(f"[bold]Email:[/] {info['contato'].get('email', 'Não informado')}")
            console.print(f"[bold]Série:[/] {info['turma'].get('serie', 'Não informado')}")
            console.print(f"[bold]Turno:[/] {info['turma'].get('turno', 'Não informado')}")

            for trimestre in ['trimestre_1', 'trimestre_2', 'trimestre_3']:
                console.print(f"\n[bold]{trimestre.replace('_', ' ').capitalize()}:[/]")
                for disciplina in info[trimestre]:
                      console.print(f"  - [bold]{disciplina['disciplina'].capitalize()}:[/] {', '.join(map(str, disciplina['provas']))}")

    except Exception as e:
        console.print(f"[red]Erro inesperado: {e}[/red]")

    input("Aperte qualquer tecla para continuar **PARE DE PROCURAR A QUALQUER TECLA \n")

inicializa_verificacao()
def PainelDeControle():
    titulo = Text("Painel de Controle da Escola", style="bold blue")

    tabela = Table(title="Informações dos Alunos")
    tabela.add_column("Matrícula", style="cyan")
    tabela.add_column("Nome", style="cyan")
    tabela.add_column("Série", style="cyan")
    tabela.add_column("Turno", style="cyan")

    alunos = abre_dados('r')

    for matricula, info in alunos.items():
        nome = info["nome"]
        serie = info["turma"]["serie"]
        turno = info["turma"]["turno"]
        if matricula == "0":
          continue
        else:
          tabela.add_row(matricula, nome, serie, turno)

    # Cria o painel com a tabela
    painel = Panel(
        tabela,
        title="Lista de Alunos",
        title_align="left",
        subtitle="Informações dos alunos matriculados",
        subtitle_align="right",
        border_style="bright_green",
        padding=(1, 2)
    )

    console.print(titulo)
    console.print(painel)

    # Simulação de interação
    console.print("\n[bold yellow]Escolha uma opção:[/bold yellow]")
    console.print("[1] Adicionar Aluno")
    console.print("[2] Alterar dados Aluno")
    console.print("[3] Mostrar Notas")
    console.print("[4] Adicionar e atualizar Notas")
    console.print("[5] Ver dados Aluno")
    console.print("[6] Total de alunos")
    console.print("[7] Boletim do aluno")
    console.print("[8] Remove Aluno")
    console.print("[9] Sair da aplicação")


    # Captura a entrada do usuário
    acao = int(console.input("\nDigite sua escolha: "))

    if acao == 1:
      adiciona_dados(alunos)

    elif acao == 2:
      altera_dados()

    elif acao == 3:
      exibir_boletim()

    elif acao == 4:
      adiciona_atualiza_nota()

    elif acao == 5:
      visualiza_dados_aluno()

    elif acao == 6:
      dados_alunos()

    elif acao == 7:
      exibir_boletim_completo()


    elif acao == 8:
     remove_aluno()

    else:
        console.print("[blue]Saindo do painel...[/blue]")
        return

while True:
    PainelDeControle()
    inicializa_verificacao()
```
